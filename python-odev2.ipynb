```python
class FilmListesi:
    def __init__(self, isim):
        self.isim = isim
        f = open(f"{isim}.txt", "a")
        f.close()

    def ekle(self, film_adi, kategori):
        f = open(f"{self.isim}.txt", "a")
        f.write(f"{film_adi}\t{kategori}\n")
        f.close()

    def ara(self, film_adi, debug=False):
        f = open(f"{self.isim}.txt", "r")
        output = f.read()
        filmler = output.split("\n")
        for index, film in enumerate(filmler):
            film = film.split("\t")
            if film_adi in film[0]:
                if not debug == True:
                    print("film bulundu! adı:{}  kategorileri:{}".format(film[0],film[1]))
                f.close()
                return index
    
    def sil(self, film_adi):
        index = self.ara(film_adi, True)
        f = open(f"{self.isim}.txt", "r")
        lines = f.readlines()
        f.close
        f = open(f"{self.isim}.txt", "w")
        for i, line in enumerate(lines):
            if not i == index:
                f.write(line+"\n")
        f.close()


liste = FilmListesi("filmler")

while True:
    print("1- ekle\n2- ara\n3- sil\n4- çıkış")
    secim = input("Lütfen bir değer giriniz")
    try:
        secim = int(secim)
    except Exception as e:
        print(f"Hata! {e.args}")
        continue

    if secim == 1:
        # ekleme işlemleri
        film_adi = input("Filmin adını giriniz")
        kategoriler = input("Film kategorilerini giriniz")
        liste.ekle(film_adi, kategoriler)
        continue
    elif secim == 2:
        # arama işlemleri
        film_adi = input("Aranacak filmin adını giriniz")
        liste.ara(film_adi)
        continue
    elif secim == 3:
        #silme işlemleri
        film_adi = input("Silinecek filmin adını giriniz")
        liste.sil(film_adi)
        continue
    elif secim == 4:
        break
    else:
        print("Hatalı seçim yaptınız")
        continue

```
